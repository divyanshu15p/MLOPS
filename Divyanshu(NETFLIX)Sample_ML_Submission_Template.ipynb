## 3. ***Data Wrangling***

### Data Wrangling Code

````python
data = data.drop_duplicates().reset_index(drop=True)

for col in data.columns:
    if data[col].dtype == "object":
        mode_value = data[col].mode()
        if len(mode_value) > 0:
            data[col] = data[col].fillna(mode_value[0])
    else:
        data[col] = data[col].fillna(data[col].median())

data["duration_num"] = data["duration"].str.extract("(\d+)").astype(float)

data["date_added"] = pd.to_datetime(data["date_added"], errors="coerce")
data["year_added"] = data["date_added"].dt.year
data["month_added"] = data["date_added"].dt.month

data["year_added"] = data["year_added"].fillna(data["year_added"].median())
data["month_added"] = data["month_added"].fillna(data["month_added"].median())

data["content_age"] = data["year_added"] - data["release_year"]
data["country_count"] = data["country"].apply(lambda x: len(str(x).split(",")))
data["genre_count"] = data["listed_in"].apply(lambda x: len(str(x).split(",")))

data.head()
````

## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
#This approach maintains distribution and avoids distortion caused by extreme values.
ml_data = data.copy()

cat_cols = ["director","cast","country","rating","listed_in"]
for col in cat_cols:
    ml_data[col] = ml_data[col].fillna(ml_data[col].mode()[0])

num_cols = ["duration_num","year_added","month_added"]
for col in num_cols:
    ml_data[col] = ml_data[col].fillna(ml_data[col].median())

ml_data.isnull().sum()
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

Missing values were handled using mode for categorical variables and median for numerical variables. This approach maintains distribution and avoids distortion caused by extreme values.

### 2. Handling Outliers

````python
cols = ["release_year","duration_num","year_added","month_added"]

for col in cols:
    q1 = ml_data[col].quantile(0.25)
    q3 = ml_data[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    before_min = ml_data[col].min()
    before_max = ml_data[col].max()

    ml_data[col] = np.where(ml_data[col] < lower, lower, ml_data[col])
    ml_data[col] = np.where(ml_data[col] > upper, upper, ml_data[col])

    after_min = ml_data[col].min()
    after_max = ml_data[col].max()

    print(col)
    print("Before → Min:", before_min, "Max:", before_max)
    print("After  → Min:", after_min, "Max:", after_max)
    print("-"*40)
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Outliers were treated using the IQR method by capping extreme values. This reduces the influence of abnormal values while preserving data structure.

### 4. Textual Data Preprocessing
(It's mandatory for textual dataset i.e., NLP, Sentiment Analysis, Text Clustering etc.)

#### 1. Expand Contraction

````python
contractions = {
    "can't": "cannot",
    "won't": "will not",
    "n't": " not",
    "'re": " are",
    "'s": " is",
    "'d": " would",
    "'ll": " will",
    "'t": " not",
    "'ve": " have",
    "'m": " am"
}

def expand_text(text):
    text = str(text)
    for key, value in contractions.items():
        text = text.replace(key, value)
    return text

ml_data["description_clean"] = ml_data["description"].apply(expand_text)
ml_data["description_clean"].head()
````

#### 2. Lower Casing

````python
ml_data["description_clean"] = ml_data["description_clean"].str.lower()
ml_data["description_clean"].head()
````

#### 3. Removing Punctuations

````python
import string

ml_data["description_clean"] = ml_data["description_clean"].str.translate(str.maketrans("", "", string.punctuation))
ml_data["description_clean"].head()
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
import re

def remove_urls_digits(text):
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\w*\d\w*", "", text)
    return text

ml_data["description_clean"] = ml_data["description_clean"].apply(remove_urls_digits)
ml_data["description_clean"].head()
````

#### 5. Removing Stopwords & Removing White spaces

````python
from nltk.corpus import stopwords
import nltk

nltk.download("stopwords")

stop_words = set(stopwords.words("english"))

def remove_stopwords(text):
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return " ".join(words)

ml_data["description_clean"] = ml_data["description_clean"].apply(remove_stopwords)
ml_data["description_clean"].head()
````

````python
ml_data["description_clean"] = ml_data["description_clean"].str.strip()
ml_data["description_clean"].head()
````

#### 6. Rephrase Text

````python
ml_data["description_clean"] = ml_data["description_clean"].str.replace(r"\s+", " ", regex=True)
ml_data["description_clean"].head()
````

#### 7. Tokenization

````python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize

ml_data["tokens"] = ml_data["description_clean"].apply(lambda x: word_tokenize(str(x)))

ml_data[["description_clean", "tokens"]].head()
````

#### 8. Text Normalization

````python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

ml_data["lemmatized_tokens"] = ml_data["tokens"].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
ml_data["final_text"] = ml_data["lemmatized_tokens"].apply(lambda x: " ".join(x))

ml_data[["final_text"]].head()
````

##### Which text normalization technique have you used and why?

Lemmatization was used to convert words into their base form, improving consistency and reducing redundancy in text data.

````python
X = model_data.drop("target", axis=1)
y = model_data["target"]

cat_cols = ["rating","country","listed_in"]
num_cols = ["release_year","duration_num","year_added","month_added","content_age","country_count","genre_count"]
text_col = "final_text"

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ("num", StandardScaler(), num_cols),
        ("text", TfidfVectorizer(max_features=3000), text_col)
    ]
)
````

## ***7. ML Model Implementation***

### ML Model - 1

````python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

model1 = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", LogisticRegression(max_iter=1000, class_weight="balanced"))
])

model1.fit(X_train, y_train)

y_pred1 = model1.predict(X_test)
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

````python

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

acc1 = accuracy_score(y_test, y_pred1)
prec1 = precision_score(y_test, y_pred1)
rec1 = recall_score(y_test, y_pred1)
f11 = f1_score(y_test, y_pred1)

metrics1 = pd.DataFrame({
    "Metric": ["Accuracy","Precision","Recall","F1 Score"],
    "Score": [acc1, prec1, rec1, f11]
})

plt.figure(figsize=(7,4))
sns.barplot(data=metrics1, x="Metric", y="Score")

plt.ylim(0.8,1)

for i, v in enumerate(metrics1["Score"]):
    plt.text(i, v + 0.01, f"{v:.3f}", ha='center')

plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
from sklearn.model_selection import GridSearchCV

param_grid1 = {
    "classifier__C": [0.1, 1, 5],
    "classifier__solver": ["liblinear", "lbfgs"]
}

grid1 = GridSearchCV(
    model1,
    param_grid1,
    cv=5,
    scoring="f1",
    n_jobs=-1
)

grid1.fit(X_train, y_train)

best_model1 = grid1.best_estimator_

print("Best Parameters:", grid1.best_params_)
print("Best CV Score:", grid1.best_score_)
````

##### Which hyperparameter optimization technique have you used and why?

GridSearchCV was used for hyperparameter optimization in Model 1. It systematically tests different combinations of parameters such as regularization strength (C) and solver types using cross-validation. This helps in selecting the best configuration based on model performance, ensuring that the Logistic Regression model is well-tuned and not relying on default parameters.

##### Have you seen any improvement? Note down the improvement with updates Evaluation metric Score Chart.

Yes, after applying hyperparameter tuning, the model showed improvement in performance. There was a slight increase in F1-score and better balance between precision and recall. The updated evaluation metric score chart reflects more consistent and stable predictions, indicating improved generalization compared to the initial baseline model.
